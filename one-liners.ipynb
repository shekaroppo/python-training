```python
import requests as r

headers = {
    'Referrer': 'https://www.bseindia.com/markets/MarketInfo/BhavCopy.aspx', 
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:128.0) Gecko/20100101 Firefox/128.0', 
    'Content-Type': 'application/octet-stream'
}

%time

bhavUrl = "https://www.bseindia.com/download/BhavCopy/Equity/BhavCopy_BSE_CM_0_0_0_20240719_F_0000.CSV"
resp = r.get(bhavUrl, headers=headers)

resp.raise_for_status()

with open("bse.csv", "wb") as f: f.write(resp.content)
```

```python
import csv

with open("bse.csv") as data:
    reader = csv.DictReader(data)
    for row in reader:
        print(row['TckrSymb'], row['ISIN'], row['ClsPric'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)

fig, ax = plt.subplots()

# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos, labels=people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('bse.csv')

# Step 1: Sort the DataFrame by 'TOTVOL' in descending order
df_sorted = df.sort_values(by='TtlTrfVal', ascending=False)

# Step 2: Select the top 20 rows
df_top20 = df_sorted.head(50)

# Step 3: Create the horizontal bar chart
plt.figure(figsize=(10, 8))
plt.barh(df_top20['TckrSymb'], df_top20['TtlTrfVal'], color='skyblue')
plt.xlabel('Total Volume')
plt.ylabel('ISIN')
plt.title('Top 20 Stocks by Total Volume')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest values on top
plt.show()
```
